```python
a=pd.read_csv("company_sales_data.csv")
x=a["month_number"].tolist()
a1=a["facewash"].tolist()
b=a["facecream"].tolist()
c=a["bathingsoap"].tolist()
d=a["shampoo"].tolist()
plt.plot(x,a1,label="month wise data",linewidth=3,marker="o",color="b")
plt.plot(x,b,label="month wise data",linewidth=3,marker="o",color="g")
plt.plot(x,c,label="month wise data",linewidth=3,marker="o",color="hotpink")
plt.plot(x,d,label="month wise data",linewidth=3,marker="o",color="pink")
plt.xlabel("total profit")
plt.ylabel("month number")
plt.xticks(x)
plt.yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])
plt.show()
```

```python
x=a["month_number"].tolist()
y=a["toothpaste"].tolist()
plt.scatter(x,y)
plt.xlabel("total profit")
plt.ylabel("number of units")
plt.xticks(x)
plt.grid(True,linewidth=1,linestyle="--")

plt.show()
```

```python
x=a["month_number"].tolist()
y=a["toothpaste"].tolist()
z=a["bathingsoap"]
plt.bar([a-0.25 for a in x],y,width=0.25,label="tootthpast",align="edge")
plt.bar([a+0.25 for a in x],z,width=-0.25,label="bathingsoap",align="edge")
plt.xlabel("month_number")
plt.ylabel("sales for month")
plt.legend(loc="upper left")
plt.title("sales data")
plt.grid(True,linewidth=1,linestyle="--")

plt.show()
```
