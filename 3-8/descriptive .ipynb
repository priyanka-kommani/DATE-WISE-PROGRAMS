### grouping and Aggregating
#### what is the average SOLD PRICE for each age category?

```python
a.groupby("AGE")["SOLD PRICE"].mean()
```

```python
soldpricebyage = a.groupby('AGE')['SOLD PRICE'].mean().reset_index() 
soldpricebyage 
```

### average SOLD PRICE for Different Playing Roles in Each Age Category?

```python
soldpricebyagerole =a.groupby(['AGE', 'PLAYING ROLE'])['SOLD PRICE'].mean()
soldpricebyagerole
```

```python
soldprice_by_age_role=a.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index() 
soldprice_by_age_role
```

### joining dataframes

### Compare the average auction price for different ages and playing roles

```python
comparison=soldprice_by_age_role.merge(soldpricebyage,on='AGE',how='outer') 
```

```python
comparison
```

####  Re-naming columns

```python
comparison.rename(columns={"AGE":"age","PLAYING ROLE":"playing role","SOLD PRICE_x":"soldprice","SOLD PRICE_y":"soldprice1"},inplace=True)
```

###  Applying Operations to multiple columns
#### percentage change in SOLD PRICE

```python
print(comparison)
```

```python
comparison['change']=comparison.apply(lambda rec:(rec.soldprice-rec.soldprice1)/rec.soldprice1,axis=1) 
```

```python
comparison
```
