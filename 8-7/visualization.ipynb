```python
import pandas as pd
df=pd.read_csv("data_set.csv")
print(df)
```

```python
import matplotlib.pyplot as plt
month=df["month_number"].tolist()
profit=df["total_profit"].tolist()
plt.plot(month,profit)
plt.xlabel("months")
plt.ylabel("profits")
plt.xticks(month)
plt.title("Company profits")
plt.show()
```

```python
plt.plot(month,profit,label="profit data of the last year",color="r",marker="o",markerfacecolor="k",
        linestyle="--",linewidth=3)
plt.xlabel("month number")
plt.ylabel("profit in doller")
plt.legend(loc="lower right")
plt.title("company sales data of the last year")
plt.xticks(month)
plt.yticks([100000,200000,300000,400000,500000])
plt.show()
```

```python
facecream=df["facecream"].tolist()
facewash=df["facewash"].tolist()
toothpaste=df["toothpaste"].tolist()
bathingsoap=df["bathingsoap"].tolist()
shampoo=df["shampoo"].tolist()
moisturizer=df["moisturizer"].tolist()
plt.plot(month,facecream,color="b",marker="o",markerfacecolor="k",label="facecream sales data",linewidth=3)
plt.plot(month,facewash,color="g",marker="o",markerfacecolor="k",label="facewash sales data",linewidth=3)
plt.plot(month,toothpaste,color="r",marker="o",markerfacecolor="k",label="toothpaste sales data",linewidth=3)
plt.plot(month,bathingsoap,color="pink",marker="o",markerfacecolor="k",label="bathingsoap sales data",linewidth=3)
plt.plot(month,shampoo,color="k",marker="o",markerfacecolor="k",label="shampoo sales data",linewidth=3)
plt.plot(month,moisturizer,color="hotpink",marker="o",markerfacecolor="k",label="moisturizer sales data",linewidth=3)
plt.legend(loc="upper left")
plt.title("sales data")
plt.xlabel("months")
plt.ylabel("sales units in number")
plt.show()
```

```python
plt.scatter(month,toothpaste,label="toothpaste sales in the month")
plt.legend(loc="upper left")
plt.title("Sales of toothpaste")
plt.xlabel("months")
plt.ylabel("number of units sold")
plt.grid(True,linestyle="--",linewidth=1)
plt.show()
```

```python
plt.bar(month,facecream,label="facecream sales",width=0.25,align="edge")
plt.bar(month,facewash,label="facewash sales",width=-0.25,align="edge")
plt.legend(loc="upper left")
plt.title("sales data")
plt.show()
```

```python
plt.bar(month,bathingsoap,label="bathingsoapsale")
plt.grid(True,linewidth=1,linestyle="--")
plt.title("sales data")
plt.legend(loc="upper right")
plt.xlabel("months")
plt.ylabel("bathingsoap sales")
plt.xticks(month)
plt.show()
```

```python
pro_range=[150000,175000,200000,225000,250000,300000,350000]
plt.hist(profit,pro_range,label="profit data")
label=["low","average","Good","Best"]
plt.legend(loc="upper left")
plt.xlabel("profit range in dollers")
plt.ylabel("actual profit in dollers")
plt.xticks(profit)
plt.title("profit data")
plt.show()
```

```python
labels=["facecream","facewah","toothpaste","bathing soap","shampoo","moisturizer"]
sales_data=[df["facecream"].sum(),df["facewash"].sum(),df["toothpaste"].sum(),df["bathingsoap"].sum(),df["shampoo"].sum(),
           df["moisturizer"].sum()]
plt.pie(sales_data,labels=labels,autopct="%1.1f%%")
plt.legend(loc="upper right")
plt.title("Sales data")
plt.axis("equal")
plt.show()
```

```python
f,sub=plt.subplots(2,sharex=True)
sub[0].plot(month,bathingsoap,label="bathingsoap sales",color="k",marker="p")
sub[0].set_title("bathing soap")
sub[1].plot(month,facewash,label="facewash sales",color="b",marker="o")
sub[1].set_title("facewash")
plt.show()
```

```python
plt.plot([],[],color="m",label="face cream",linewidth=5)
plt.plot([],[],color="c",label="face wash",linewidth=5)
plt.plot([],[],color="r",label="tooth paste",linewidth=5)
plt.plot([],[],color="g",label="bathing soap",linewidth=5)
plt.plot([],[],color="b",label="moisturizer",linewidth=5)
plt.plot([],[],color="k",label="shampoo",linewidth=5)
plt.stackplot(month,facecream,facewash,toothpaste,bathingsoap,moisturizer,shampoo,
             colors=["m","c","r","g","b","k"])
plt.title("All products sales")
plt.xlabel("months in number")
plt.ylabel("sales units in number")
plt.legend(loc="upper left")
plt.show()
```
